```python
from sqlalchemy import create_engine, inspect
from pathlib import Path

db_path = Path("users.db").resolve()  # note a barra normal
if not db_path.exists():
    raise FileNotFoundError(f"Banco não encontrado: {db_path}")

engine = create_engine(f'sqlite:///{db_path}')
insp = inspect(engine)
tabelas = insp.get_table_names()

print("Tabelas encontradas no banco:")
for t in tabelas:
    print("-", t)
```

```python
import sqlite3
import pandas as pd

# Caminho do banco SQLite
db_path = "users.db"

# Conexão
conn = sqlite3.connect(db_path)

# Query com LEFT JOIN
query = """
SELECT
    u.*,
    i.image_b64
    
FROM user_registration u
LEFT JOIN images i
    ON u.infopen = i.infopen
"""

# Carrega no DataFrame
df = pd.read_sql_query(query, conn)

# Fecha conexão
conn.close()

# Visualizar
df.head()
```
